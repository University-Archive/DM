```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import LabelBinarizer
```

```python
student_df = pd.read_csv("./data/student.csv", sep=';')
student_df
```

```python
student_df.isnull().sum()
```

```python
student_df.drop(["reason", "guardian"], axis=1, inplace = True)
```

```python
student_df.loc[student_df["school"] == "GP" ,"school"] = 1
student_df.loc[student_df["school"] == "MS" ,"school"] = 0
student_df["sex"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["sex"]))
student_df["address"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["address"]))
student_df["famsize"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["famsize"]))
student_df["Pstatus"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["Pstatus"]))
student_df["Mjob"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["Mjob"]))
student_df["Fjob"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["Fjob"]))
student_df["schoolsup"] = pd.DataFrame(LabelBinarizer().fit_transform(student_df["schoolsup"]))
student_df
```

```python
train, test = train_test_split(student_df, test_size=0.2)
train
```

```python
plt.scatter(train.index, train["G3"])
```
