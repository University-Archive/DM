```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
covid_df = pd.read_csv("./data/covid.csv")
covid_df
```

```python
birth_year = covid_df["birth_year"]
print("mean: ", birth_year.mean())
print("max: ", birth_year.max())
print("std: ", birth_year.std())
```

```python
covid_df.isnull().sum()
```

```python
covid_df.drop(["infection_reason", "infected_by"], axis=1, inplace = True)
covid_df.dropna(inplace = True)
covid_df
```

```python
plt.scatter(covid_df["id"], covid_df["birth_year"])
plt.ylabel("Birth Year")
plt.show()
```

```python
plt.hist(covid_df["birth_year"])
plt.xlabel("Birth Year")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.matshow(covid_df[["birth_year", "birth_year"]], aspect='auto')
plt.show()
```
